その日にもらう米粒の数のグラフ

```python
#リストに保存する
days_count = []
i = 1 
for count in range(100):
     i = i * 2
     days_count.append(i)

import matplotlib.pyplot as plt
%matplotlib inline
# グラフを表示する。タイトルと軸ラベル名付き。
plt.title("Daily get count")
plt.xlabel("days")
plt.ylabel("number of rice")
plt.plot(days_count)  # 「リスト名」のところにリストの変数名を入れる
plt.show()
```

## 【問題2】100日目以外の米粒の数

```python
 def sorori_shinzaemon(n_days=100):
    """曽呂利新左衛門の米の逸話における、日にちごとの米の粒の数を計算する関数。
    Parameteres
    -----------
    n_days: int
        米を貰う日数 (default : 100)
    Returns
    -------
    list_n_grains : list
        ある日に貰う米の数のリスト (リストの長さはn_daysに等しい)
    list_total_grains : list
        ある日までに累計で貰う米の数のリスト (リストの長さはn_daysに等しい)
     """
    list_n_grains = [] #　ある日に貰う米の数の空のリスト
    days = 1
    n = 2 ** (n_days -1 )  
    list_n_grains.append(n)
    
    print("{}日目にもらう米の数は{}粒です。".format(n_days,list_n_grains))  #ある日にもらう米の数
    
    list_total_grains = [] #rある日までに累計で貰う米の数の空のリスト
    days = 1 　#その日にもらう米の数
    days2 = 0　#累計の数
    for count in range(n_days):
         days = days * 2
         days2 = days2 + days
         list_total_grains.append(days2)
     
    print("{}日目までにもらう米の累計の数は{}粒です。".format(n_days,list_total_grains[-1]))
   
#日数を入力する
n_days = 90   #コース受講日を90日とする
sorori_shinzaemon(n_days)
```

## 【問題3】貰える米で何人が何日生活できるか

```python
def days_of_llife(number_of_people,grain):
    """この米があれば何人が何日間生活できるかを計算する関数。
    Parameteres
    -----------
    number_of_people: int
         生活する人数
    grain: int
         米粒の数
    -------
    total_grains : list
        　何日間で貰える米粒の数のリスト
         
　　※　成人一人の1日の米の消費量
  　　年間の1人当たりの米の消費量　53.5㎏　（農林水産省ホームページより）
    　1日の消費量＝　53.5　÷　365（日）≒　150ｇ
      グラムを粒に換算　　米は1000粒で約22.5ｇ　150　÷　22.5　×　1000　≒　6666（粒）
      今回は1人当たり1日の消費量を6666粒で計算する
     """
    total_grains = [] #ある日までに累計で貰う米の数の空のリスト
    days = 1   # days == １日でもらう米の数
    total = 0  # total == これまでにもらった米の累計の数
    for count in range(grain):
         days = days * 2
         total = total + days
         total_grains.append(total)
            
    possible_days = total_grains[-1] // (6666 * number_of_people)
    print("{}日間分の米で{}人が生活できる日数は{}日です。".format(grain,number_of_people,possible_days))

#米がもらえる日数を入力する
days = 90
#生活する人数を入力する
number_of_people = 5
days_of_llife(number_of_people,days)
```

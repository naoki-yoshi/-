# ダミーデータの作成

# 【問題1】乱数の作成

```python
from matplotlib import  pyplot as plt
import numpy as np
```

```python
np.random.seed(0)
```

```python
number = 500
cov = [[1,0.8],[0.8,1]]
mean = [-3,0]

datal = np.random.multivariate_normal(mean,cov,number)
print("作成したデータのshape:{}".format(datal.shape))
```

```python
datal
```

# 【問題2】散布図による可視化

```python
plt.title("scatter")
plt.xlabel("x1")
plt.ylabel("x2")
plt.scatter(datal[:,0],datal[:,1])
plt.show()
```

#  【問題3】ヒストグラムによる可視化

```python
plt.hist(datal[:,0])
plt.xlim(-6,0)
```

```python
plt.hist(datal[:,1])
plt.xlim(-3,3)
```

# 問題4】データの追加

```python
number = 500
cov = [[1,0.8],[0.8,1]]
mean = [0,-0.3]

data2 = np.random.multivariate_normal(mean,cov,number)
```

```python
data2
```

```python
plt.title("scatter")
plt.xlabel("x1")
plt.ylabel("x2")
plt.scatter(datal[:,0],datal[:,1],label=0)
plt.scatter(data2[:,0],data2[:,1],label=1)
plt.legend()
plt.show()
```

# 【問題5】データの結合

```python
a_ndarray = np.concatenate([datal,data2],0)
```

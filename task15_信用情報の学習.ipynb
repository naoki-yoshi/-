```python
#bureau_balanceを集約する
bureau_balance_sum = bureau_balance.groupby("SK_ID_BUREAU").sum()
```

```python
#
bureau_merge = pd.merge(bureau,bureau_balance_sum, on="SK_ID_BUREAU",how="left" )
```

```python
#各データのSK_ID_CURRを集約する
POS_CASH_balance_sum = POS_CASH_balance.groupby("SK_ID_CURR").sum()
credit_card_balance_sum =credit_card_balance.groupby("SK_ID_CURR").sum()
installments_payments_sum =installments_payments.groupby("SK_ID_CURR").sum()
previous_application_sum =previous_application.groupby("SK_ID_CURR").sum()
bureau_merge_sum = bureau_merge.groupby("SK_ID_CURR").sum()
```

```python
# インデックスをSK_ID_CURRにする
application_train_dummies = application_train_dummies.set_index('SK_ID_CURR')
```

```python
# 不要なカラムを削除する
bureau_merge_sum = bureau_merge_sum.drop('SK_ID_BUREAU', axis=1)
POS_CASH_balance_sum = POS_CASH_balance_sum.drop('SK_ID_PREV', axis=1)
credit_card_balance_sum = credit_card_balance_sum.drop('SK_ID_PREV', axis=1)
previous_application_sum = previous_application_sum.drop('SK_ID_PREV', axis=1)
installments_payments_sum = installments_payments_sum.drop('SK_ID_PREV', axis=1)
```

```python
# 結合後どこのデータかわかりやすいように接頭語付与
bureau_merge_sum.columns = ["BUREAU_" + column for column in bureau_merge_sum.columns]
POS_CASH_balance_sum.columns = ["POS_" + column for column in POS_CASH_balance_sum.columns]
credit_card_balance_sum.columns = ["CREDIT_" + column for column in credit_card_balance_sum.columns]
previous_application_sum.columns = ["APPLICATION_" + column for column in previous_application_sum.columns]
installments_payments_sum.columns = ["INSTALLMENTS_" + column for column in installments_payments_sum.columns]
```

####application_train_dummiesにbureau_balance以外を結合する

```python
data = pd.concat([application_train_dummies,
    bureau_merge_sum,
    POS_CASH_balance_sum,
    credit_card_balance_sum,
    previous_application_sum,
    installments_payments_sum
],axis=1)
data = data.sort_index(axis=1)
```

```python
# 中央値による欠損値補完
data_median = data.fillna(data.median())
```

訓練データと検証データに分轄

```python
# 説明変数のカラム取得
columns = list(data_median.columns)
columns.remove("TARGET")#dataからTARGETを抜き出す
columns
# 分割
X_train, X_test, y_train, y_test = train_test_split(data_median.loc[:,columns], data_median["TARGET"], test_size=0.3,  random_state=0)
```

## 標準化する

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

### 訓練データを使った学習

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier()
clf.fit(X_train, y_train)
```

```python
# 予測値の算出
prediction = clf.predict(X_train)
prediction
```

```python
# 混同行列計算
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_train, prediction)
tn, fp, fn, tp = cm.flatten()

print("True Negative：",tn)
print("False Positive：",fp)
print("False Negative：",fn)
print("True Positive：",tp)
```

###テストデータを用いた評価

```python
# 予測値の算出
prediction = clf.predict(X_test)
# 混同行列計算
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, prediction)
tn, fp, fn, tp = cm.flatten()

print("True Negative：",tn)
print("False Positive：",fp)
print("False Negative：",fn)
print("True Positive：",tp)
```

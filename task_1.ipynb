問題１　べき乗の算術演算子を使用した実装

```python
THICKNESS = 0.00008
folded_thickness = THICKNESS *(2 ** 43)
print("厚さ： {}メートル".format(folded_thickness))
```

問題２　単位の変換　

```python
print("厚さ： {:.2f}キロメートル".format(folded_thickness / 1000))
```

月までの距離はおよそ 384,400 km　紙の厚さは703687.44㎞　

問題　３　for文を活用して計算

```python
for count in range(43):
    THICKNESS = THICKNESS * 2

print("厚さ： {}メートル".format(THICKNESS))
```

問題4 計算時間の比較

```python
# べき乗の算術演算子を使用した実装
import time
start = time.time()
#####
THICKNESS = 0.00008
folded_thickness = THICKNESS *(2 ** 43)
print("厚さ： {}メートル".format(folded_thickness))
#####
elapsed_time = time.time() - start
print("time : {}[s]".format(elapsed_time))
```

```python
#  for文を活用して計算
import time
start = time.time()
#####
for count in range(43):
    THICKNESS = THICKNESS * 2

print("厚さ： {}メートル".format(THICKNESS))
#####
elapsed_time = time.time() - start
print("time : {}[s]".format(elapsed_time))
```

問題5 リストへの保存

```python
process_thickness = []
i = 0.00008
for count in range(44):
    i = i * 2
    process_thickness.append(i)

print(len(process_thickness))
```

問題6 折れ線グラフの表示

```python
import matplotlib.pyplot as plt
%matplotlib inline
# グラフを表示する。タイトルと軸ラベル名付き。
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness [m]")
plt.plot(process_thickness)  # 「リスト名」のところにリストの変数名を入れる
plt.show()
```

※　厚さの変化は初期の頃は小さく、後半にかけて急激に厚さが増している。

　問題７　グラフのカスタマイズ

```python
#アレンジ１　線を赤色に
import matplotlib.pyplot as plt
%matplotlib inline
# グラフを表示する。タイトルと軸ラベル名付き。
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness [m]")
plt.plot(process_thickness, color = 'red')  # 「リスト名」のところにリストの変数名を入れる
plt.show()
```

```python
#アレンジ２　緑色の点線　折った回数の個所にポイント
import matplotlib.pyplot as plt
%matplotlib inline
# グラフを表示する。タイトルと軸ラベル名付き。
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness [m]")
plt.plot(process_thickness, color='green', marker='o', linestyle='dashed',)  # 「リスト名」のところにリストの変数名を入れる
plt.show()
```

```python
#　アレンジ２の太線
import matplotlib.pyplot as plt
%matplotlib inline
# グラフを表示する。タイトルと軸ラベル名付き。
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness [m]")
plt.plot(process_thickness, 'go--', linewidth=2, markersize=12,)  # 「リスト名」のところにリストの変数名を入れる
plt.show()
```

まとめ
折り紙を４３回折り曲げるという問いに対して、べき乗の算術演算子を用いる場合とfor文を持ちいる場合の２通りで答えを求めた。
紙を折り曲げていくと最初のうちは大した厚みの変化はないが、後半になると一回の折り曲げで大きく変化することが、グラフにより視覚的にもわかりやすいものになっている。
